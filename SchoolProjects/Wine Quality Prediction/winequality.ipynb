```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
# Load the dataset
df = pd.read_csv('winequality.csv')
```

```python
# Drop unnecessary columns (like 'Id' if present)
if 'Id' in df.columns:
    df.drop('Id', axis=1, inplace=True)
```

```python
# Check and fill null values with the column mean
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
# Sepearate features and target
X = df.drop('quality', axis = 1)
Y = df['quality']
```

```python
# Split into training and testing set
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Create LinearRegression model

model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
# Predict on test set
Y_pred = model.predict(X_test)
```

```python
# Evaluate the model
mse = mean_squared_error(Y_test, Y_pred)
print("Coefficients: ", model.coef_)
print("Intercept: ", model.intercept_)
print("Mean Squared Error: ", mse)
```

```python
# Predict quality of a new wine sample
new_sample = np.array([[8, 0.4, 0.40, 15, 0.048, 40, 150, 0.99, 3, 0.45, 10.5]])
predicted_quality = model.predict(new_sample)
print("Predicted wine quality: ", predicted_quality[0])
```
